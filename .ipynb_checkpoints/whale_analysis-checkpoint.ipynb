## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx = pd.read_csv('sp_tsx_history.csv', index_col='Date', parse_dates=True, infer_datetime_format=True)
```

```python
# Check Data Types
sp_tsx.dtypes
```

```python
# Fix Data Types - removed $ sign and , so it can changed to a float
sp_tsx['Close'] =sp_tsx['Close'].str.replace('$','', regex = True)
sp_tsx['Close'] = sp_tsx['Close'].str.replace(',','', regex = True)
#sp_tsx['Close'] = pd.to_numeric(sp_tsx['Close'])
```

```python
# changed close values to float for further calculating the percent change
sp_tsx[['Close']] = sp_tsx[['Close']].astype(float)
sp_tsx.dtypes
# Calculate Daily Returns
sp_tsx['pct_chng'] = sp_tsx['Close'].pct_change()
```

```python
# check for null values 
#sp_tsx.isna().sum()
# Drop nulls
sp_tsx.dropna(inplace = True)
#verify if number of records changed to confirm null was dropped 
#sp_tsx
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_2 = sp_tsx.rename(columns={"pct_chng": "S&P TSX"})
sp_tsx_2.drop('Close', 1, inplace = True)
```

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios
whale_r.plot(legend = True)
algo_r.plot(legend = True)
sp_tsx_2.plot(legend = True)
#calculate and plot for the remaining portfolios 
portfolio = [l_hist, otex_hist, shop_hist]
for value in portfolio:
    value['pct_chng'] = value['Close'].pct_change()
    value.dropna(inplace = True)
    value.drop('Close', 1, inplace = True)
l_hist_2 = l_hist.rename(columns={"pct_chng": "L Historical"})
l_hist_2.plot(legend = True)
otex_hist_2 = otex_hist.rename(columns={"pct_chng": "OTEX Historical"})
otex_hist_2.plot(legend = True)
shop_hist_2 = shop_hist.rename(columns={"pct_chng": "SHOP Historical"})
shop_hist_2.plot(legend = True)
#combine all the portfolios to calculate comulative returns in the next step
mast_data = [whale_r,algo_r,sp_tsx_2,l_hist_2,otex_hist_2,shop_hist_2]
mast_data_2 = pd.concat(mast_data, axis ="columns", join="inner")
```

#### Calculate and Plot cumulative returns.

```python
mast_data_2 = mast_data_2.drop('Symbol',1)

# Calculate cumulative returns of all portfolios
cum_return = (mast_data_2+1).cumprod()
# Plot cumulative returns
cum_return.plot(figsize=(15, 10))
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
#mast_data = [whale_r,algo_r,sp_tsx_2,l_hist_2,otex_hist_2,shop_hist_2]
whale_r_1 = whale_r.rolling(window =21).std()
algo_r_1 = algo_r.rolling(window =21).std()
l_hist_1 = l_hist_2.rolling(window =21).std()
otex_hist_1 = otex_hist_2.rolling(window =21).std()
shop_hist_1 = shop_hist_2.rolling(window =21).std()
# Plot the rolling standard deviation
whale_r_1.plot()
algo_r_1.plot()
l_hist_1.plot()
otex_hist_1.plot()
shop_hist_1.plot()
```

### Calculate and plot the correlation

```python
# Calculate the correlation

mast_data_2.corr()

# Display de correlation matrix

import seaborn as sns
sns.heatmap(mast_data_2.corr(), vmin = -1, vmax = 1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
cov_w = whale_r['SOROS FUND MANAGEMENT LLC'].cov(whale_r['SOROS FUND MANAGEMENT LLC'])
# Calculate variance of S&P TSX
sp_var = sp_tsx_2['S&P TSX'].var()
# Computing beta
beta = cov_w /sp_var

# Plot beta trend
rolling_covariance = whale_r['SOROS FUND MANAGEMENT LLC'].rolling(window=60).cov(whale_r['SOROS FUND MANAGEMENT LLC'])
rolling_variance = sp_tsx_2['S&P TSX'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance 
rolling_beta.plot(figsize=(20,10), title = 'Rolling 60 Days Beta')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (mast_data_2.mean() * 252) / (mast_data_2.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar()
```
